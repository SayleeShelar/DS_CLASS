```python
Class=['a','b','c','d']
a_marks=[80,75,90,63]
plt.bar(Class, a_marks, color = "blue")
plt.xlabel("class")
plt.ylabel("marks")
plt.title("class records")
plt.ylim(50) # y axis will start from 50
plt.grid(color='green')
plt.show()
```

```python
p = np.linspace(50,100,50)
p
# linspace(start, end, no. of samples to generate)
```

```python
plt.figure(figsize=(5,2.7))
```

```python
plt.figure(figsize=(5,2.7))
plt.plot(p, p*1.5, label ='linear')
plt.plot(p, p**2, label = 'quadratic')
plt.xlabel('x label (p)')
plt.ylabel('y label(operations on P)')
plt.title("Simple Plot")
plt.legend
```
